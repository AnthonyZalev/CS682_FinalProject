```python
import matplotlib.pyplot as plt
import numpy as np

resnet50_acc = [.7711, 0.7519, 0.7236, .6602, .4077]
vicreg_sim_acc = [0.7642, 0.7503, 0.7392, 0.7035, 0.5096]
vicreg_seq_acc = [0.7876, 0.7677, 0.7527, 0.7095, 0.6193]
splits = ['80-20', '85-15', '90-10', '95-5', '99-1']

for i in range(len(splits)):
    plt.figure(figsize=(8, 5))

    plt.bar('ResNet50', resnet50_acc[i], color='blue', label='ResNet50')
    plt.bar('Sim VICReg Representations', vicreg_sim_acc[i], color='red', label='Simultaneously Trained VICReg')
    plt.bar('Seq VICReg Representations', vicreg_seq_acc[i], color='green', label='Sequentially Trained VICReg')

    plt.xlabel('Model')
    plt.ylabel('Accuracy')
    plt.title(f'Accuracy Comparison ({splits[i]} Unlabeled-Labeled Data Split)')
    plt.legend()
    plt.grid(True)

    plt.show()
```
